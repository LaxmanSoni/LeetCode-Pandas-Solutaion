# Task 1 Create a DataFrame from List
Write a solution to create a DataFrame from a 2D list called student_data. This 2D list contains the IDs and ages of some students.

The DataFrame should have two columns, student_id and age, and be in the same order as the original 2D list.

The result format is in the following example.

```python
import pandas as pd



student_data = [
    [1, 20],
    [2, 22],
    [3, 21],
    [4, 23],
    [5, 19]
]


columns = ['student_id', 'age']

df = pd.DataFrame(student_data, columns=columns)

print(df)
```

# Task 2  Get the Size of a DataFrame

Write a solution to calculate and display the number of rows and columns of players.

Return the result as an array:

[number of rows, number of columns]


```python
import pandas as pd 

players = {
    "Player_id":[1,2,3],
    "Name":["Laxman","viki","ravi"],
    "Age":[20,21,22]
  
}

df = pd.DataFrame(players)
rows , columns = df.shape
result  = [rows , columns]
print(df)
print("Numer of row and coulumn :", result)
```

```python
temperature_data = pd.DataFrame({'city': ['Jacksonville']*5 + ['ElPaso']*5,
                                 'month': ['January', 'February', 'March', 'April', 'May']*2,
                                 'temperature': [13, 23, 38, 5, 34, 20, 6, 26, 2, 43]})


pivoted_data = temperature_data.pivot(index='month', columns = 'city', values='temperature').reset_index()

pivoted_data
```

```python
sales_data = pd.DataFrame({'product': ['Umbrella', 'SleepingBag'],
                           'quarter_1': [417, 800],
                           'quarter_2': [224, 936],
                           'quarter_3': [379, 93],
                           'quarter_4': [611, 875]})

melted_data = pd.melt(sales_data , id_vars=['product'], var_name='quarter',value_name ='sales')


melted_data
```

```python
animals = pd.DataFrame({'name': ['Tatiana', 'Khaled', 'Alex', 'Jonathan', 'Stefan', 'Tommy'],
                        'species': ['Snake', 'Giraffe', 'Leopard', 'Monkey', 'Bear', 'Panda'],
                        'age': [98, 50, 6, 45, 100, 26],
                        'weight': [464, 41, 328, 463, 50, 349]})

result = (
    animals[animals['weight']>100].sort_values(by='weight', ascending=False)['name'].tolist()
)
result
```
